```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
sns.set_context('talk')
```

```python
df = pd.read_csv('model accuracy.csv', header = 0)
```

```python
fig, ax = plt.subplots(1,3, figsize=(24,6))

df_1 = df[df['Model'] == 'Clinical']
sns.swarmplot(data = df_1, x="Category", y="Value", hue="Dataset", ax = ax[0], dodge= True, color='black', alpha = 0.5,
              size = 10)
t = sns.barplot(data = df_1, x="Category", y="Value", hue="Dataset", ci='sd', 
                    palette="dark", alpha=.6, capsize=.1, ax=ax[0])
ax[0].set(ylim=(0.6,1), ylabel = 'Concordance', xlabel='Model', title = 'Clinical')
ax[0].legend()
for p in t.patches:
    t.annotate(format(p.get_height(), '.3f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 30), 
                   textcoords = 'offset points') 

df_2 = df[df['Model'] == 'Clinical + TIICs']
sns.swarmplot(data = df_2, x="Category", y="Value", hue="Dataset", ax = ax[1], dodge= True, color='black', alpha = 0.5,
             size = 10)
t = sns.barplot(data = df_2, x="Category", y="Value", hue="Dataset", ci='sd', 
                    palette="dark", alpha=.6, capsize=.1, ax=ax[1])
ax[1].set(ylim=(0.6,1), ylabel = 'Concordance', xlabel='Model', title = 'Clinical + TIICs')
ax[1].legend()
for p in t.patches:
    t.annotate(format(p.get_height(), '.3f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 30), 
                   textcoords = 'offset points') 

df_3 = df[df['Model'] == 'Clinical + Genes']
sns.swarmplot(data = df_3, x="Category", y="Value", hue="Dataset", ax = ax[2], dodge= True, color='black', alpha = 0.5,
             size = 10)
t = sns.barplot(data = df_3, x="Category", y="Value", hue="Dataset", ci='sd', 
                    palette="dark", alpha=.6, capsize=.1, ax=ax[2])
ax[2].set(ylim=(0.6,1), ylabel = 'Concordance', xlabel='Model', title = 'Clinical + Genes')
ax[2].legend(loc=2)
for p in t.patches:
    t.annotate(format(p.get_height(), '.3f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 30), 
                   textcoords = 'offset points') 

plt.tight_layout()
fig.savefig('test.svg')
```
